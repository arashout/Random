```python
%matplotlib inline
from sympy.solvers import solve
from sympy import Symbol, plot
```

```python
p = Symbol('p')
q = 1000 - 4*p
c_p = 24
c_f = 12875
r = p*q - (c_p*q + c_f)
```

```python
ans = solve(r, p)
for a in ans:
    print(a.evalf())
    print(q.subs(p, a).evalf())
```

```python
plot(r, (p, 0, 250))
```
